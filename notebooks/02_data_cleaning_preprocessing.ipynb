# Data Cleaning & Preprocessing
This notebook handles all missing value imputation for the raw dataset. Each column with missing values is addressed individually using an appropriate strategy — mode for categorical columns and median for numerical columns. A missing-value indicator is also created for `Credit_History` before imputation to preserve the missingness signal.

## 1. Importing Libraries

```python
import pandas as pd
import numpy as np
```

## 2. Load Raw Dataset

```python
df_raw = pd.read_csv("../data/loan_data_set.csv")
```

## 3. Create Working Copy
Creating a copy of the raw data to preserve the original untouched. All cleaning operations are performed on `df_clean`.

```python
df_clean = df_raw.copy()
```

## Credit_History
### Credit History Missing Indicator
Creating a binary indicator column to flag missing Credit History values.
- `1` = Credit History is missing (NaN)
- `0` = Credit History is present

```python
df_clean["Credit_History_Missing"] = df_clean["Credit_History"].isna().astype(int)
```

```python
df_clean["Credit_History_Missing"].value_counts()
```

### Verification
- Value counts: distribution of missing vs present
- Since Missing values in Credit_History was 50, there is also 50 1s in credit_History_Missing

### Imputation — Credit History
Filling missing values in `Credit_History` with **mode** since it is a categorical column.

```python
df_clean["Credit_History"] = df_clean["Credit_History"].fillna(df_clean["Credit_History"].mode()[0])
```

```python
# Verify no missing values remain
df_clean.isnull().sum()
```
