##  Married

```python
df_clean.isnull().sum()
```

#### Details Before
- Missing rows in Married = 3
- Imputation method = Mode

```python
df_clean["Married"] = df_clean["Married"].fillna(df_clean["Married"].mode()[0])
```

```python
df_clean.isnull().sum()
```

#### Details After imputing
- After Cleaning missing in Married Column = 0

### GENDER

```python
df_clean["Gender"].isnull().sum()
```

#### Before 
- Missing Gender = 13
- Type = Nominal Categorical
- Method = Mode

```python
df_clean["Gender"] = df_clean["Gender"].fillna(df_clean["Gender"].mode()[0])
```

```python
df_clean["Gender"].isnull().sum()
```

#### After
- Missing value = 0
- Number of rows = 614

## Dependents, Self_Employed 

```python
df_clean["Dependents"].isnull().sum()
```

```python
df_clean["Self_Employed"].isnull().sum()
```

#### Before
- Dependents Missing = 15
- Self_Employed Missing = 32
- Strategy = Mode
- Type = Categorical

```python
for col in ["Dependents","Self_Employed"]:
    df_clean[col] = df_clean[col].fillna(df_clean[col].mode()[0])
```

```python
df_clean["Dependents"].isnull().sum()
```

```python
df_clean["Self_Employed"].isnull().sum()
```

#### After
- Missing for both = 0
- Rows are not affected

## LoanAmount, Loan_Amount_Term

```python
df_clean["LoanAmount"].isnull().sum()
```

```python
df_clean["Loan_Amount_Term"].isnull().sum()
```

#### Before
- LoanAmount Missing = 22
- Loan_Amount_Term Missing = 14
- Strategy = Median
- Type = Numerical

```python
for col in ["LoanAmount","Loan_Amount_Term"]:
    df_clean[col]=df_clean[col].fillna(df_clean[col].median())
```

```python
df_clean["Loan_Amount_Term"].isnull().sum()
```

```python
df_clean["LoanAmount"].isnull().sum()
```

#### After
- LoanAmount Missing = 22
- Loan_Amount_Term Missing = 14
- Numbers of rows uneffected
